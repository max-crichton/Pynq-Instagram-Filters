# PS Image Filtering

The library used to perform filtering using the PS is PIL thanks to its multiple functions which correspond to the ones we will use on the hardware

```python
from PIL import ImageFilter
from PIL import Image
```

Firstly, the image to be filtered is imported and shown below.

```python
img = Image.open('assets/peppers.bmp')
img
```

## Filtering using software

Having done that, three filters will be used on the image:
* Sharpen
* Blur
* Edge Detection

The timing of the filter will also be measured from the duration of the filter function (before the image is displayed) for all 3 examples.

### Sharpen Filter

```python
%%timeit
sharp_image = img.filter(ImageFilter.SHARPEN)
```

```python
sharp_image = img.filter(ImageFilter.SHARPEN)
sharp_image
```

### Blur Filter

```python
%%timeit
blurred_image = img.filter(ImageFilter.BLUR)
```

```python
blurred_image = img.filter(ImageFilter.BLUR)
blurred_image
```

### Edge Detection Filter

```python
%%timeit
edge_image = img.filter(ImageFilter.FIND_EDGES)
```

```python
edge_image = img.filter(ImageFilter.FIND_EDGES)
edge_image
```
